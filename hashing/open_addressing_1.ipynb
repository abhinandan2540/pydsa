```python
class MyHash:
    def __init__(self, m:int):
        self.hash_t=[-1]*m
        self.m=m

    def insert(self,num:int):
        index=num%self.m 
        if self.hash_t[index]==-1:
            self.hash_t[index]=num
            
        else:
            for i in range(1,self.m):
                k=(index+i)%self.m  # linear probing for collission
                if self.hash_t[k]==-1:
                    self.hash_t[k]=num
                    break
                if self.hash_t[k]==num:
                    break
        return self.hash_t
    
    def remove(self, num:int):
        i=self.hash_t.index(num) # can also be done manually
        self.hash_t.remove(num)
        self.hash_t[i]=-2
        return self.hash_t
    
    def search(self, num:int):
        return num in self.hash_t
    
```

```python
h=MyHash(m=7)
print(h.insert(49))
print(h.insert(56))
print(h.insert(72))
print(h.search(56))
print(h.remove(56))
print(h.search(56))
```
