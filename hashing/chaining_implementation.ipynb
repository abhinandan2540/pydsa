chaining is one way of handelling collision in hashtable

```python

class myhash:
    def __init__(self, bucket:int):
        self.bucket=bucket
        self.hashlist=[[] for _ in range(self.bucket)]

    def insert(self, num:int):
        index=num%self.bucket
        self.hashlist[index].append(num)
        return self.hashlist
    
    def remove(self,num:int):
        index=num%self.bucket
        if num in self.hashlist[index]:
            self.hashlist[index].remove(num)
            return self.hashlist
        else:
            return -1
        
    def search(self, num:int):
        index=num%self.bucket
        return num in self.hashlist[index]
    
```

```python
h=myhash(bucket=7)
print(h.insert(70))
print(h.insert(71))
print(h.insert(56))
print(h.insert(9))
print(h.search(56))
print(h.remove(56))
print(h.search(56))
```

```python
# normally 
# time complexity: O(1), wrost case O(n)
# space complexity: O(n+b), n is number to be inserted, b is the number of bucket in hashtable
```
