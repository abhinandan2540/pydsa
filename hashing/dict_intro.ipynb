```python
# for key to get the value, use .get()
d = {110: 'abc', 101: 'xyz', 105: 'pqr'}
print(d.get(110))
print(d.get(105))
print(d.get(125)) # if nothing present then .get() returns None
# but normal method returns error
print(d[125]) # returns KeyError
```

```python
# removing element in dictionary
d = {110: 'abc', 101: 'xyz', 105: 'pqr', 106: 'bcd'}
print(d.pop(105)) # .pop(key) delete key,val pair and return what it's deleted

# update key, value pair
d[101]="wxy"
print(d)

del d[106] # delete key 106 with corresponding value
print(d)

d[108]='cde'
print(d)
print(d.popitem()) # .popitem() delets last element and returns also
```
