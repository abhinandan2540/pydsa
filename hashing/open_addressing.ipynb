**open addressing**

open addressing handel collision with a single array only.

for open addressing, `no of slots in hash table >= no of keys to be inserted`

`linear probing, quadratic probing, double hashing`

```python
def linearProbling(arr:list, m:int):
    hash_table=[-1]*m
    for num in arr:
        index=num%m
        if hash_table[index]==-1:
            hash_table[index]=num
        else:
            k=index
            count=0
            while(count<m and hash_table[k]!=-1 and hash_table[k]!=num): # conflicting condition
                k=(k+1)%m
                count+=1
            if (count<m and hash_table[k]==-1):
                hash_table[k]=num
    return hash_table
```

```python
arr=[4, 14, 24, 44]
m=10
print(linearProbling(arr=arr, m=m))
```

```python
# with for loop (preferred)
def linearProbing(arr, m):
    hash_t=[-1]*m
    for num in arr:
        index=num%m 
        if hash_t[index]==-1:
            hash_t[index]=num
            continue
        
        if hash_t[index]==num: # dublicate element
            continue
        else:
            for i in range(1,m):
                k=(index+i)%m 

                if (hash_t[k]==-1):
                    hash_t[k]=num
                    break
                if (hash_t[k]==num):
                    break
    return hash_t
```

```python
arr = [4, 14, 24, 44]
m = 10
print(linearProbling(arr=arr, m=m))
```

```python
def quadraticProbing(arr:list, m:int):
    hash_t=[-1]*m
    for num in arr:
        index=num%m 
        if hash_t[index]==-1:
            hash_t[index]=num
            continue
        
        if hash_t[index]==num:
            continue
        else:
            for i in range(1,m):
                k=(index+i*i)%m 
                if (hash_t[k]==-1):
                    hash_t[k]=num
                    break
                if (hash_t[k]==num):
                    break
    return hash_t
```

```python
arr = [21, 10, 33, 43]
m=5
print(quadraticProbing(arr=arr, m=m))
```
