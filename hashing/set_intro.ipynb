```python
# adding element in set
s={10,20}
s.add(30)
s.add(30) # dublicates are not allowed in set
print(s)

# updating set
s.update([40,50])
print(s)

# inserting multiple list
s.update([60,70],[80,90])
print(s)
```

```python
# removing element in set
s={10,20,30,40}
print(s)

s.remove(30)
print(s)

s.discard(20)
print(s)

s.clear() # clean the entire set
print(s)

del s  # delete the entire set object
s.add(30)
print(s)
```

```python
# operation of two sets
s1={2,4,6,8}
s2={3,6,9}

# union of sets
# for union time complexity: O(len(s1)+len(s2))
print(s1|s2)
print(s1.union(s2))

# intersection of sets
# for intersection, time complexity: O(min(len(s1), len(s2)))
print(s1&s2)
print(s1.intersection(s2))

# difference
# time complexity depends on the respective set
# for 1st, time complexity O(len(s1))
# for 2nd, time complexity O(len(s2))
print(s1-s2) # present in s1 but not in s2
print(s2-s1) # present in s2 but not in s1


# symmetric difference
# time complexity: O(len(s1)+len(s2))
print(s1^s2) # not present in both
```

```python
# some important operations
s1={2,4,6,8}
s2={4,8}

# disjoint set: not anything between common
print(s1.isdisjoint(s2)) # 4,8 common

# subset: a set containing only elements that are also present in
# another large set
print(s1.issubset(s2)) # s1 is the subset of s2: False
print(s2.issubset(s1)) # s2 is the subset of s1: True

# another method for defining subset
print(s1<=s2) # s1 is the subset of s2
print(s2<=s1) # s2 is the subset of s1

# superset
# a set is superset when all the element of other set is present in
# the larger set
print(s1.issuperset(s2)) # s1 is superset of s2
print(s2.issuperset(s1)) # s2 is the superset of s1

# proper subset
print(s1<s2) # s1 is propet subset of s2

# proper superset
print(s1>s2) # s1 is proper superset of s2

```
