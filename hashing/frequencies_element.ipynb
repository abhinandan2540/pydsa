```python
def frquencies_element(arr:list):
    hash={}
    for i in range(len(arr)):
        count=1
        if arr[i] not in hash.keys():
            hash[arr[i]]=count
        else:
            hash[arr[i]]+=count
    return hash
```

```python
arr=[10,12,10,15,10,20,12,12]
print(frquencies_element(arr=arr))
```
