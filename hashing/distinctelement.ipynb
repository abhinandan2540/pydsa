```python
# count distinct element in array
# using hash
def distinctElement(arr:list):
    arr_hash={}
    for i in range(len(arr)):
        count=1
        if arr[i] not in arr_hash:
            arr_hash[arr[i]]=count
        else:
            arr_hash[arr[i]]+=count
    return len(arr_hash.keys())
```

```python
arr=[10,20,10,30,30,20]
print(distinctElement(arr=arr))
```

```python
# using set
def distinctElement(arr:list):
    s=set(arr)
    return len(s)
```

```python
arr = [10, 20, 10, 30, 30, 20]
print(distinctElement(arr=arr))
```
