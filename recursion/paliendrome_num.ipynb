```python
# method 1
def num_paliendrome(n:int):
    c_num=n
    n_num=0
    while (n>0):
        l_d=n%10
        n_num=n_num*10+l_d
        n=n//10
    if(n_num==c_num):
        return 1
    else:
        return 0
    
```

```python
print(num_paliendrome(1223))
```

```python
# method 2
def num_paliendrome(n:int):
    return check_num(n=n, c_num=n, n_num=0)


def check_num(n:int, c_num:int, n_num:int):
    if(n==0):
        if (c_num==n_num):
            return 1
        else:
            return 0
        
    n_num=n_num*10+(n%10)
    n=n//10
    return check_num(n, c_num, n_num)
```

```python
print(num_paliendrome(1221)) # time complexity: O(logn)
```
