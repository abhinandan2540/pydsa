```python
# Recursion basic example

def func1():
    print('func1() called')

def func2():
    print("Before func1()")
    func1()
    print("After func1()")

print("Before func2()")
func2()
print("After func2()")
```

A function is called recursive, when it called itself directly or indirectly

```python
# it'll be an infinite recursion call
def func():
    print("dsawithpy")
    return func()

func()
```

```python
# controlling recursion
def func(n:int):
    # base case matters most for stopping recursion
    if(n<=0):
        return 
    print("dsawithpy")
    return func(n-1)

func(5)
```
