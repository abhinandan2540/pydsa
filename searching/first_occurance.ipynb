```python
def first_occurance(nums:list, n:int):
    high=len(nums)-1
    low=0
    while(low<=high):
        mid=(low+high)//2
        if(nums[mid]>n):
            high=mid-1
        elif(nums[mid]<n):
            low=mid+1
        else: # nums[mid]==n
            if(mid==0 or nums[mid-1]!=nums[mid]): # check if same element exists before
                return mid
            else:
                high=mid-1
    return -1
    
```

```python
nums=[1, 2,2,2,3,3,3,4]
print(first_occurance(nums=nums, n=2)) # O(logn)
```
