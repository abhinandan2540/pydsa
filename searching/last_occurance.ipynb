```python
def last_occurance(nums:list, n:int):
    low=0
    high=len(nums)-1
    while(low<=high):
        mid=(low+high)//2
        if(nums[mid]>n):
            high=mid-1
        elif(nums[mid]<n):
            low=mid+1
        else:
            if(mid==len(nums)-1 or nums[mid]!=nums[mid+1]):
                return mid
            else:
                low=mid+1
    return -1
```

```python
nums = [1,2,2,2,3,3,3,3,4]
print(last_occurance(nums=nums, n=3)) # O(logn)
```
