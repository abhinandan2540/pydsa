```python
# find floor in a sorted array

def findFloor(arr:list, k:int):
    low=0
    high=len(arr)-1
    res=-1
    while(low<=high):
        mid=(low+high)//2
        if(arr[mid]==k):
            return mid
        elif (arr[mid]<k):
            res=mid
            low=mid+1
        else:
            high=mid-1

    return res

nums=[1,2,8,10,10,12,19]    
# print(findFloor(nums,11))
print(findFloor(nums,10))
```

```python
# applying the last occurance
def lastOccurance(nums:list, x:int):
    low=0
    high=len(nums)-1
    while(low<=high):
        mid=(low+high)//2
        if(nums[mid]>x):
            high=mid-1
        elif(nums[mid]<x):
            low=mid+1
        else:
            if(mid==len(nums)-1 or nums[mid]!=nums[mid+1]):
                return mid
            else:
                low=mid+1
    return -1


nums = [1, 2, 8, 10, 10, 12, 19]
print(lastOccurance(nums,10))
```

```python
# the floor

def floorNum(arr:list,x:int):
    low=0
    high=len(arr)-1
    res=-1
    while(low<=high):
        # mid=(low+high)//2
        mid=low+(high-low)//2
        if(arr[mid]<=x):
            res=mid
            low=mid+1
        else:
            high=mid-1
    return res


nums = [1, 2, 8, 10, 10, 12, 19]
print(floorNum(nums,10))
```
