```python
# there's another method for finding counting of 1's

def count_occurance(arr:list):
    n=len(arr)
    low=0
    high=n-1
    while(low<=high):

        mid=(low+high)//2
        if((arr[mid]==n-1 or arr[mid+1]==0) and arr[mid]==1):
            return mid+1
            # for left 
        elif (arr[mid]==0):
            high=mid-1
            # for right
        elif(arr[mid]==1):
            low=mid+1
    return 0

arr=[1,1,1,1,0,0,0]
print(count_occurance(arr))        
```

```python
# method 1
def first_occurance(nums:list):
    low=0
    high=len(nums)-1
    while(low<=high):
        mid=(low+high)//2
        if(nums[mid]>1):
            high=mid-1
        elif(nums[mid]<1):
            low=mid+1
        else:
            if(mid==0 or nums[mid-1]!=nums[mid]):
                return mid
            else:
                high=mid-1
    return -1

def last_occurance(nums:list):
    low=0
    high=len(nums)-1
    while (low <= high):
        mid = (low+high)//2
        if (nums[mid] > 1):
            high = mid-1
        elif (nums[mid] < 1):
            low = mid+1
        else:
            if (mid == len(nums)-1 or nums[mid] != nums[mid+1]):
                return mid
            else:
                low=mid+1
    return -1


def count_occurance(nums:list):
    if(first_occurance(nums=nums)==-1):
        return 0
    else:
        return last_occurance(nums=nums)-first_occurance(nums=nums)+1
    
```

```python
nums = [0,0,0,1,1,1,1]
print(count_occurance(nums=nums)) # O(logn)
```

```python
# method 2
# condition: array must be sorted in non-decreasing order
def count_ones(nums:list):
    low=0
    high=len(nums)-1
    while(low<=high):
        mid=(low+high)//2
        if(nums[mid]==1 and (mid==len(nums)-1 or nums[mid+1]==0)): # A
            return mid+1
        elif(nums[mid]==0):
            high=mid-1
        elif(nums[mid]==1):
            low=mid+1
    return -1

# A: 
# arr[mid]==1: if the current element is 1 
# and
# (mid==len(nums)-1 or nums[mid+1]==0): either mid is last element or next element is 0
# meaning: this 1 is the final 1 in the array 
# then
# return mid+1
```

```python
nums = [1,1,1,1,0,0,0]
print(count_ones(nums=nums)) # O(logn)
```
