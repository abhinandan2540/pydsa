```python
# ceil of a number
def findCeil(arr: list, x: int):
    low = 0
    high = len(arr)-1
    res = -1
    while (low <= high):
        # mid=(low+high)//2
        mid = low+(high-low)//2
        if (arr[mid] >= x):
            res = mid
            low = mid+1
        else:
            high = mid-1
    return res


nums = [1, 2, 8, 10, 10, 12, 19]
print(findCeil(nums, 10))

# time complexity : o(lon(n))
```
