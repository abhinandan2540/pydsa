*Binary Search works well for sorted array*

```python
# method 1
def binary_search(arr:list, n:int):
    for i in range(len(arr)):
        if arr[i]==n:
            return i
    return -1
```

```python
arr = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
print(binary_search(arr=arr, n=5)) # O(n)
```

```python
# method 2
def binary_search(arr:list, n:int):
    low=0
    high=len(arr)-1
    while(low<=high):
        mid=(low+high)//2
        if(arr[mid]==n):
            return mid
        elif(arr[mid]>n):
            high=mid-1
        else:
            low=mid+1
    return -1
```

```python
arr = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
print(binary_search(arr=arr, n=38)) # O(logn)
```
