```python
def first_occurance(nums: list, k: int):
    low = 0
    high = len(nums)-1
    while (low <= high):
        mid = (low+high)//2
        if (nums[mid] > k):
            high = mid-1
        elif (nums[mid] < k):
            low = mid+1
        else:
            if (mid == 0 or nums[mid] != nums[mid-1]):
                return mid
            else:
                high = mid-1
    return -1


def last_occurance(nums: list, k: int):
    low = 0
    high = len(nums)-1
    while (low <= high):
        mid = (low+high)//2
        if (nums[mid] > k):
            high = mid-1
        elif (nums[mid] < k):
            low = mid+1
        else:
            if (mid == len(nums)-1 or nums[mid] != nums[mid+1]):
                return mid
            else:
                low = mid+1
    return -1


def count_occurance(nums:list,n:int):
    if(first_occurance(nums=nums, k=n)==-1):
        return 0
    else:
        return last_occurance(nums=nums, k=n)-first_occurance(nums=nums, k=n)+1
    
```

```python
nums = [10, 20, 20, 20,20, 30, 30]
print(count_occurance(nums=nums, n=30)) # O(logn)
```
