```python
# method 1
# square root of a number
import math
def sqr_root(n:int):
    return math.floor(n**0.5)

print(sqr_root(8))
```

```python
# method 2
def square_root(n:int):
    low=1
    high=n
    ans=-1
    while(low<=high):
        mid=(low+high)//2
        mid_sqr=mid*mid
        if(mid_sqr==n):
            return mid
        elif(mid_sqr>n):
            high=mid-1
        else: # (mid^2<n)
            low=mid+1
            ans=mid # when (mid^2<n), mid is the feasible solution, store it in ans before moving further
    return ans 
```

```python
print(square_root(26)) # O(logn)
```
