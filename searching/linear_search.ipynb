```python
# method 1
def linear_search(nums:list, n:int):
    for i in range(len(nums)):
        if(nums[i]==n):
            return i
    return -1
```

```python

nums = [0, 20, 80, 30, 60, 50,
        110, 100, 130, 170]
print(linear_search(nums=nums, n=30))
```

```python
# method 2
# using two pointer approach
def linear_search(nums:list, n:int):
    left=0
    right=len(nums)-1

    while left<=right:
        if(nums[left]==n):
            return left
        if (nums[right]==n):
            return right
        left+=1
        right-=1

    return -1
```

```python
nums=[10,20,30,40,50]
print(linear_search(nums=nums, n=20))
```
