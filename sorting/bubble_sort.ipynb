Bubble sort: 
simplest sorting algorithm, it's stable and in-place.

Basic of Bubble sort is: it chooses the large value and put it in the end position of the array and keep doing it 
                         untill it become sorted array.

```python
# method 1
def bubble_sort(nums:list):
    for i in range(len(nums)-1):
        for j in range(len(nums)-i-1):
            if(nums[j]>nums[j+1]):
                nums[j], nums[j+1]=nums[j+1], nums[j]
    return nums
```

```python
nums = [10,8,20,5]
print(bubble_sort(nums=nums)) # O(n^2)
```

```python
# method 2
# what if array is sorted from middle

def bubble_sort(nums:list):
    sorted_nums=sorted(nums)
    for i in range(len(nums)-1):
        for j in range(len(nums)-i-1):
            if (nums[j] > nums[j+1]):
                nums[j], nums[j+1] = nums[j+1], nums[j]
        if (nums==sorted_nums):
            return nums
    return nums
```

```python
nums = [8, 5,2,7,9,10]
print(bubble_sort(nums=nums))
```

```python
# method 3
def bubble_sort(nums:list):
    for i in range(len(nums)-1):
        swapped=False
        for j in range(len(nums)-i-1):
            if (nums[j] > nums[j+1]):
                nums[j], nums[j+1] = nums[j+1], nums[j]
                swapped=True
        if swapped==False:
            return nums
    return nums

    
```

```python
nums = [8, 5, 2, 7, 9, 10]
print(bubble_sort(nums=nums)) # O(n) if array sorted else O(n^2)
```
