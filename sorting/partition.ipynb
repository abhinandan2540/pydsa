**partition**

let say we want every small element (less than 8) in one side and every other in another

```python
# method 1
def naive_parititon(nums:list, p:int):
    n=len(nums)
    nums[p],nums[n-1]=nums[n-1],nums[p]
    result=[]
    for i in range(n):
        if(nums[i]<=nums[n-1]):
            result.append(nums[i])
    for i in range(n):
        if(nums[i]>nums[n-1]):
            result.append(nums[i])
    return result
```

```python
nums = [5,13,6,9,12,8,11]
print(naive_parititon(nums=nums,p=3)) # O(n), space complexity: O(n)
```

```python
# method 2
def lomuto_parition(nums:list, p:int):
    """p:partition number index """
    n=len(nums)
    nums[p],nums[n-1]=nums[n-1],nums[p]
    pivot=nums[n-1]
    i=-1
    for j in range(n-1): # excluding pivot element from operation
        if(nums[j]<pivot):
            i+=1
            nums[i],nums[j]=nums[j],nums[i]
    nums[i+1],nums[n-1]=nums[n-1],nums[i+1]
    return nums
```

```python
nums=[5, 13,6,9,12,8,11]
print(lomuto_parition(nums=nums,p=3))
# time complexity: O(n)
# space complexity: O(1)
```

```python
# method 3
def hoare_partition(nums:list, low:int, high:int, p:int):
    """p: index of pivot number"""
    pivot=nums[p]
    i=low-1
    j=high+1
    while True:
        i+=1
        while(nums[i]<pivot):
            i+=1
        j-=1
        while(nums[j]>pivot):
            j-=1
        if(i>=j):
            return j # returns the index number of the partition
        nums[i],nums[j]=nums[j],nums[i]

```

```python
nums=[5, 13,6,9,12,8,11]
print(hoare_partition(nums=nums, low=0, high=6, p=2))
# time complexity: O(n)
# space complexity: O(1)
```
