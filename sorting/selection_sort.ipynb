Selection sort: 
unstable, in-place for sorting in computation.

Basic Idea: it chooses the smallest element first and put it in the first of an array.
            keep doing it, untill array become sorted.

Time complexity : O(n^2)
space complexity : O(1)

```python
def selection_sort(nums:list):
    n=len(nums)
    for i in range(n-1):
        min_indx=i
        for j in range(i+1,n):
            if(nums[j]<nums[min_indx]):
                min_indx=j
        nums[min_indx],nums[i]=nums[i],nums[min_indx]
    return nums
```

```python
nums = [64, 25, 12, 22, 11]
print(selection_sort(nums=nums))  
# time complexity: O(n^2)
# space complexity: O(1)
```
