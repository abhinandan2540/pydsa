Selection sort: 
unstable, in-place for sorting in computation.

Basic Idea: it chooses the smallest element first and put it in the first of an array.
            keep doing it, untill array become sorted.

Time complexity : O(n^2)
space complexity : O(1)

```python
def selection_sort(nums:list):
    for i in range(len(nums)-1):
        min_idx=i
        for j in range(i+1,len(nums)):
            if(nums[j]<nums[min_idx]):
                min_idx=j
        # first it search the entire array, find smallest one then swap
        nums[min_idx], nums[i] = nums[i], nums[min_idx]
    return nums
```

```python
nums=[64, 25,12,22,11]
print(selection_sort(nums=nums)) # o(n^2)
```
