```python
# soretd: also used for sorting, it's applicable on iterables. eg: set, dict, tuple, list, etc
# sort : modified the new list
# sorted: returns a new modificiation list twin

l1=[2,3,1]
l2=sorted(l1)
print(l2)

l3=[10,-15,-2,1]
l4=sorted(l3, key=abs) # abs: absolute (mod) value
print(l4)

l5=sorted(l3, key=abs, reverse=True)
print(l5)
```

```python
# sorted manipulation
# working with tuple
t=(10,12,5,1)
t1=sorted(t)
# output: list
print(t1)

# working with set
s={'dsa','with','py'}
s1=sorted(s)
print(s1)

# working with string
my_str='DSA'
str_2=sorted(my_str)
print(str_2)

# working with dict
d={1:'a',3:'b',2:'c'}
d1=sorted(d)
print(d1)
```
