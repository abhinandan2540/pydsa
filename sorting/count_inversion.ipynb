**what is count inversion**

An inversion count in an array measures how far or close an array is from being sorted. It represents the total number of pairs of elements a[i],a[j] such that (i<j) but (a[i]>a[j]). 

A sorted array has 0 inversions, while a reverse-sorted array has the maximum number of inversions.Â 

```python
# method 1
def count_inversion(nums:list):
    inversion=0
    n=len(nums)
    if(nums==sorted(nums, reverse=True)):
        return (n*(n-1))//2
    else:
        for i in range(n):
            for j in range(i+1,n):
                if(nums[i]>nums[j]):
                    inversion+=1
    return inversion
```

```python
nums=[2, 4,1,3,5]
print(count_inversion(nums=nums)) # O(n^2)
```

```python
# method 2
# modified merge sort algorithm for count inversion
def count_merge_subarray(nums:list, low:int, high:int, mid:int):
    nums1=nums[low:mid+1]
    nums2=nums[mid+1:high+1]
    m=len(nums1)
    n=len(nums2)
    i=j=0
    k=low
    count=0
    while(i<m and j<n):
        if(nums1[i]<=nums2[j]):
            nums[k]=nums1[i]
            i+=1
        else:
            nums[k]=nums2[j]
            j+=1
            count+=(m-i) # count will add if and only if nums2[j]'s is less than nums1[i]'s
            # where nums2[j] and nums1[i] are sorted array
            # if nums2[0] is less than nums1[0] then automatically it less than all the elements of nums1[i]'staticmethod
            
        k+=1
    while(i<m):
        nums[k]=nums1[i]
        i+=1
        k+=1
    while(j<n):
        nums[k]=nums2[j]
        j+=1
        k+=1
    return count


def count_inversion(nums:list, low:int, high:int):
    count=0
    if(high>low):
        mid=(low+high)//2
        count+=count_inversion(nums=nums, low=low,high=mid)
        count+=count_inversion(nums=nums, low=mid+1, high=high)
        count+=count_merge_subarray(nums=nums, low=low, high=high, mid=mid)
    return count

```

```python
nums = [5, 8, 11, 14, 2, 7]
print(count_inversion(nums=nums, low=0, high=len(nums)-1))  # o(nlogn)
```
