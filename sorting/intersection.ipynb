```python
# method 1
def intersection_array(nums1:list, nums2:list):
    m=len(nums1)
    n=len(nums2)
    result=set()
    if(m>n):
        for i in range(m):
            if(nums1[i] in nums2):
                result.add(nums1[i])
    else:
        for i in range(n):
            if(nums2[i] in nums1):
                result.add(nums2[i])
    return list(result)
```

```python
nums1 = [1,1,3,3]
nums2 = [1,1,3,5,7]
print(intersection_array(nums1=nums1, nums2=nums2)) #O(m.n)
```

```python
# á¹ƒethod 2
def intersection_array(nums1:list, nums2:list):
    nums2=set(nums2) # O(n)
    result=set()
    for i in nums1: # O(m)
        if i in nums2:
            result.add(i)
    return list(result)
```

```python
nums1 = [1, 1, 3, 3]
nums2 = [1, 1, 3, 5, 7]
print(intersection_array(nums1=nums1, nums2=nums2)) # O(m+n)
```

```python
# method 3
def intersection_array(nums1:list, nums2:list):
    m=len(nums1)
    n=len(nums2)
    i=0
    j=0
    result=[]
    while(i<m and j<n):
        if(i>0 and nums1[i]==nums2[i-1]):
            i+=1
            continue
        if(nums1[i]==nums2[j]):
            result.append(nums1[i])
            i+=1
            j+=1
        elif (nums1[i]<nums2[j]):
            i+=1
        else:
            j+=1

    return result
```

```python
nums1 = [1, 1, 3, 3]
nums2 = [1, 1, 3, 5, 7]
print(intersection_array(nums1=nums1, nums2=nums2)) # O(m+n)
```
