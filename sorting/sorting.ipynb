```python
# python has inbuilt sort method for sorting

l1=[4,2,10,9]
l1.sort()
print(l1)

l2 = [4, 2, 10, 9]
l2.sort(reverse=True)
print(l2)

l3=['dsa','with','py']
l3.sort()
print(l3)
```

```python
# sort manipulation

def myFunc(s):
    return len(s)

l1=['dsa','with','py']
# sorting with respect to len
l1.sort(key=myFunc)
print(l1)

l2 = ['dsa', 'with', 'py']
l2.sort(key=myFunc,reverse=True)
print(l2)
```
