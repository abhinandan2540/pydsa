quick sort using *lomuto's algorithm*

```python
def lomuto_partition(nums:list, low:int, high:int):
    pivot=nums[high] # pivot always the last element for quick sort implemenation
    i=low-1

    for j in range(low,high):
        if(nums[j]<pivot):
            i+=1
            nums[i],nums[j]=nums[j],nums[i]
    nums[i+1],nums[high]=nums[high],nums[i+1]
    return i+1 


def quicksort_lomuto(nums:list, low:int, high:int):
    if(high>low):
        p=lomuto_partition(nums=nums, low=low, high=high)
        quicksort_lomuto(nums=nums, low=low, high=p-1)
        quicksort_lomuto(nums=nums, low=p+1, high=high)
    return nums
    
```

```python
nums = [5, 13, 6,9,12,8,11]
print(quicksort_lomuto(nums=nums, low=0, high=6))
# best & avergae cases
# time complexity: O(nlogn)
# space complexity: O(logn) # quick sort is recursive
# wrost case
# time complexity: O(n^2)
# space complexity: O(n)
```
