**Stable & Unstable sort**

stable sort: *in which the relative position of input and output is same. e.g. bubble sort, insertion sort, merge sort etc.*

unstable sort: *in which the relative position of input and output is not same. e.g. selection sort, quick sort, heap sort.*

```python
nums = [(3, 'A'), (1, 'B'), (3, 'C'), (2, 'D')]
nums.sort()
print(nums) # stable sort
# before sorting (3,A),..(3,C)
# after sorting (3,A),..(3,C)

# but for unstable sorting it will be like (3,C),..(3,A)
```

**in-place & not in-place sort**

in-place: *if the algorithm dose not take any extra space for computation. e.g. bubble sort, insertion sort, selection sort, quick sort.*

not in-place: *take extra space for sorting. e.g. merge sort, counting sort etc.*
