**hoare algorithm**

```python
def hoare_algorithm(nums:list, low:int, high:int, p:int):
    """p: pivot number index"""
    i=low-1
    j=high+1
    pivot=nums[p]
    while True:
        i+=1
        while(nums[i]<pivot):
            i+=1
        j-=1
        while(nums[j]>pivot):
            j-=1
        if(i>=j):
            return j # returns the index number of pivot after sucessfull partition

        nums[i],nums[j]=nums[j],nums[i]
```

```python
nums=[5, 13, 6, 9, 12, 8, 11]
print(hoare_algorithm(nums=nums, low=0, high=6,p=3))
# time complexity: O(n)
# space complexity: O(1)
```
