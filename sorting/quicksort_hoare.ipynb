quick sort using *hoare's algorithm*

```python
def hoare_partition(nums:list, low:int, high:int):
    pivot=nums[low]
    i=low-1
    j=high+1
    
    while True:
        i+=1
        while(nums[i]<pivot):
            i+=1
        j-=1
        while(nums[j]>pivot):
            j-=1
        if(i>=j):
            return j
        
        nums[i],nums[j]=nums[j],nums[i]

def quicksort_hoare(nums:list, low:int, high:int):
    if(high>low):
        p=hoare_partition(nums=nums, low=low, high=high)
        quicksort_hoare(nums=nums, low=low, high=p)
        quicksort_hoare(nums=nums, low=p+1, high=high)
    return nums

```

```python
nums = [5, 13, 6, 9, 12, 8, 11]
print(quicksort_hoare(nums=nums, low=0, high=6))
# best and avg case
# time complexity: O(nlogn)
# space complexity: O(logn)
# wrost case
# time complexity: O(n^2)
# space complexity: O(n)
```
