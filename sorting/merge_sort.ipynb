Merge Sort: stable, not-in place algorithm

```python

def merge_subarray(nums: list, low: int, high: int, mid: int):
    nums1 = nums[low:mid+1]
    nums2 = nums[mid+1:]
    m = len(nums1)
    n = len(nums2)
    i = 0
    j = 0
    k = low
    while (i < m and j < n):
        if (nums1[i] <= nums2[j]):
            nums[k] = nums1[i]
            i += 1
            k += 1
        else:
            nums[k] = nums2[j]
            j += 1
            k += 1

    while (i < m):
        nums[k] = nums1[i]
        i += 1
        k += 1

    while (j < n):
        nums[k] = nums2[j]
        j += 1
        k += 1
    return nums
```

```python
def merge_sort(nums, low, high):
    if(high>low):
        mid=(low+high)//2
        merge_sort(nums,low, mid)
        merge_sort(nums,mid+1,high)
        merge_subarray(nums,low,high,mid)
    return nums
```

```python
nums=[10, 5, 30, 15, 7]
print(merge_sort(nums=nums,low=0,high=4)) # T: O(nlogn), S:O(n)
```
