```python
# method 1
def union_array(a:list, b:list):
    res=[]
    m=len(a)
    n=len(b)
    i=0
    j=0
    while(i<m and j<n):
        if (a[i]<=b[j]):
            if(a[i] not in res):
                res.append(a[i])
            i+=1
        else:
            if (b[j] not in res):
                res.append(b[j])
            j+=1

    while(i<m):
        if (a[i] not in res):
            res.append(a[i])
        i+=1

    while(j<n):
        if(b[j] not in res):
            res.append(b[j])
        j+=1

    return res
```

```python

a = [2, 20, 20, 40]
b = [1, 10, 20]
print(union_array(a, b)) # O(m+n)
```

```python
# method 2
def union_array(nums1:list, nums2:list):
    result=set()
    m=len(nums1)
    n=len(nums2)
    i=0
    j=0
    while(i<m and j<n):
        if(nums1[i]<nums2[j]):
            result.add(nums1[i])
            i+=1
        else:
            result.add(nums2[j])
            j+=1
    while(i<m):
        result.add(nums1[i])
        i+=1
    while(j<n):
        result.add(nums2[j])
        j+=1
    return list(result)
```

```python
a = [2, 20, 20, 40]
b = [1, 10, 20]
print(union_array(a, b))  # O(m+n)
```
