**lomuto partition**

```python
def lomuto_algorithm(nums:list, p:int):
    """p: pivot number index"""
    pivot=nums[p]
    n=len(nums)
    nums[p],nums[n-1]=nums[n-1],nums[p]
    i=-1
    for j in range(n-1):
        if (nums[j]<pivot):
            i+=1
            nums[i],nums[j]=nums[j],nums[i]
    nums[i+1],nums[n-1]=nums[n-1],nums[i+1]
    return nums
```

```python
nums = [5,13,6,9,12,8,11]
print(lomuto_algorithm(nums=nums, p=3))
# time complexity: O(n)
# space complexity: O(1)
```
