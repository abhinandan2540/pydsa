In this problem, we have to print all thepossible prime numbers upto n numbers

```python
#solution 1

#making a function for prime numbers
def isPrime(n:int):
    is_prime=True
    if(n==1):
        is_prime=False
    if (n==2 or n==3):
        return is_prime
    if(n%2==0 or n%3==0):
        is_prime=False
    i=5
    while(i*i<=n):
        if(n%i==0 or n%(i+1)==0):
            is_prime=False
        i+=6
    return is_prime

# isPrime(63)
# isPrime(7)


# making a function for returning all the prime numbers
def seive_of_eratosthenes(n:int):
    for i in range(2,n+1):
        if (isPrime(i)):
            print(i, end=" ")
      

# seive_of_eratosthenes(10)
seive_of_eratosthenes(23)


# time complexity : O(n.n^1/2)
```
