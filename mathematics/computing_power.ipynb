we have to calcuate power of a certain number

```python
def compute_power(x:int, n:int): # x:base, n:power
    if(n==0):
        return 1
    return x*compute_power(x,n-1)

compute_power(5,2) # time complexity O(n)
```

```python

def n_power(x,y):
    if(y==0):
        return 1
    temp=n_power(x,y//2)
    if(y%2==0):
        return temp*temp
    else:
        return x*temp*temp
    
print(n_power(4,3)) # time complexity: O(logn)
```

```python
def num_power(x:int, n:int):
    ans=1
    while(n>0):
        if(n%2!=0):
            ans*=x
        x*=x
        n=n//2
    return ans
```

```python
print(num_power(2,5)) # time complexity : O(logn)
```
