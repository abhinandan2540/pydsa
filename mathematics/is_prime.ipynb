```python
def is_prime(n:int):
    isPrime=True
    if(n==1):
        isPrime=False
    if(n==2 or n==3):
        return isPrime
    if (n%2==0 or n%3==0):
        isPrime=False
    i=5
    while(i*i<=n):
        if(n%i==0) or (n%(i+2)==0):
            isPrime=False
        i+=6
    return isPrime
```

```python
print(is_prime(11)) # time complexity: O(n^1/2)
```
