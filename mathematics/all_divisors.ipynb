fidning all divisors of a number

```python
def n_divisors(n:int):
    for i in range(1,n+1):
        if (n%i==0):
            print(i)

n_divisors(10) # time complexity: O(n)
```

```python
def n_divisors_2(n:int):
    i=1
    while (i*i<=n):
        if (n%i==0):
            print(i)
            if(i!=n//i):
                print(n//i)
        i+=1

n_divisors_2(10) # time complexity: O(n^1/2)       
```

```python
# additional
def n_divisors_3(n:int):
    i=1
    while (i*i<=n):
        if(n%i==0):
            print(i,end=" ")
        i+=1
    i-=1
    while (i>=1):
        if(n%i==0):
            print(n//i,end=" ")
        i-=1

n_divisors_3(30)
# time complexity : O(n^1/2) 
```
