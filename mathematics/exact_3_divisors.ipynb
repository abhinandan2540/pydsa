```python
# exact 3 divisors
# finding a number having exact 3 divisors
# -> find a prime number, cause prime number square has exact 3 divisors
# eg: 3^2=9 (1,2,9) etc


def is_prime(n:int):
    isPrime=True
    if (n==1):
        isPrime=False
    if (n==2 or n==3):
        return isPrime
    if (n%2==0 or n%3==0):
        isPrime=False
    i=5
    while (i*i<=n):
        if(n%i==0) or (n%(i+2)==0):
            isPrime=False
        i+=6
    return isPrime
```

```python
def exact_3_divisors(n:int):
    count=0
    i=2
    while(i*i<=n):
        if is_prime(i): # time complexity: O(n^1/2)
            count+=1
        i+=1
    return count
```

```python
print(exact_3_divisors(50)) # time complexity: O(n^1/2)
```
