```python
# number of trailing zero in a factorial depends on number of 5 in that number
# number of 0 in a factorial of a number = floor(n/5) + floor(n/25) + floor(n/125) +.....

def trailing_zero(n:int):
    t_o=0
    i=5
    while (n//i>=1):
        t_o+=n//i
        i*=5
    return t_o
```

```python
print(trailing_zero(5))
```
