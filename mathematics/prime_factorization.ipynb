```python
# prime factorization
# finding prime -> factorization

def is_prime(n:int):
    isPrime=True
    if (n==1):
        isPrime=False
    if(n==2 or n==3):
        return isPrime
    if(n%2==0 or n%3==0):
        isPrime=False
    i=5
    while(i*i<=n):
        if(n%i==0 or n%(i+2)==0):
            isPrime=False
        i+=6
    
    return isPrime
```

```python
def prime_factorization(n:int):
    for i in range(2,n+1):
        if is_prime(i): # time complexity: O(n^1/2)
            x=i
            while (n%x==0):
                print(i,end=" ")
                x*=i
```

```python
prime_factorization(100) # time complexity: O(n^2)
```
