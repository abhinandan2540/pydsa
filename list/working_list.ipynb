* working with list in python
-----------------------------------------------------

* Advantage of list container
----------------------------------------
1. List data structure got random access of elements
by providing index number we can get the respected element
2. another important functionality of list is 'cache friendliness'
cache : a small memory
CPU have lot's of cache, when we need some data, CPU first check the cache, it the data is present it gives us, else go towards RAM
this cache, reduces memory usage in much value.

List has this same concept,
when we take some data from the list, CPU not only fetches that data (element), but aslo prefetches some of the
extra element from the list too.
so when we make next call, it gives fast result

as example
l=[10,20,30,40,50]
when we call l[0]
CPU also prefetches l[1],l[2],l[3] etc
so next operation became easy.

* Disadvantages of list container
------------------------------------------------
1. insertion, deletion is slow
in insertion, deletion, we have to move back, move forward every item, so it's much slower. (take linear time O(n) )

2. searching is also slow
if list not sorted, it'll take max time.



so in list, append() and pop() only works in constant time O(1), else others like
insert(), remove(), work in O(n) time.
