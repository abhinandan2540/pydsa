```python
def list_reverse(nums:list):
    if(len(nums)<=1):
        return nums
    else:
        nums=nums[::-1]
    return nums
```

```python
nums=[1,2,3,4,5]
print(list_reverse(nums=nums)) # time complexity: O(n)
```

```python
def list_reverse(nums:list):
    n=len(nums)
    s=0
    e=n-1
    while s<e:
        nums[s],nums[e]=nums[e],nums[s]
        s+=1
        e-=1
    return nums
```

```python
nums = [1, 2, 3, 4, 5]
print(list_reverse(nums=nums))  # time complexity: O(n)
```
