```python
# method 1
def second_largest(nums:list):
    max_1=max(nums)
    max_2=None
    n=len(nums)
    if (n<=1):
        return None
    else:
        for i in nums:
            if (i!=max_1):
                if max_2==None:
                    max_2=i
                else:
                    max_2=max(i, max_2)
    return max_2
    
```

```python
nums=[1,6,3,4,5]
print(second_largest(nums=nums)) # time complexity: O(n)
```

```python
# method 2
def second_largest(nums:list):
    max_1=nums[0]
    max_2=None
    if(len(nums)<=1):
        return None
    else:
        for i in nums[1:]:
            if (i>max_1):
                max_2=max_1
                max_1=i
            elif (i!=max_1):
                if(max_2 is None or max_2<i):
                    max_2=i
    return max_2
```

```python
nums = [1, 6, 3, 4, 5]
print(second_largest(nums=nums))  # time complexity: O(n)
```
