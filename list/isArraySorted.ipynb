```python
# method 1
def isSorted(nums:list):
    n=len(nums)
    sorted=True
    if(n<=1):
        return sorted
    else:
        for i in range(n-1):
            if (nums[i]>=nums[i+1]):
                sorted=False
            else:
                return sorted
    return sorted
    
```

```python
# nums = [10,20,20,30,40,50]
# nums=[10]
nums=[]
print(isSorted(nums=nums))
```

```python
# method 2: optimal approach
def isSorted(nums:list):
    incFlag=True
    decFlag=True
    n=len(nums)
    if(n<=1):
        return 1
    else:
        for i in range(n-1):
            if (nums[i]>nums[i+1]):
                incFlag=False
            if (nums[i]<nums[i+1]):
                decFlag=False
            if not incFlag and not decFlag:
                return 0
    return 1
```

```python
# nums=[10,20,30]
# nums=[30,20,10]
nums=[30,10,20]
print(isSorted(nums=nums)) # time complexity: O(n)
```
