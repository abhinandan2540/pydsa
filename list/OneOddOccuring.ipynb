```python
# method 1
def one_odd_occuring(nums:list):
    occ_dict={}
    count=1
    odd_occ=None
    for i in range(len(nums)):
        if nums[i] in occ_dict.keys():
            occ_dict[nums[i]]+=count
        else:
            occ_dict[nums[i]]=count
    for k,v in occ_dict.items():
        if(v%2!=0):
            odd_occ=k
    return odd_occ
            
```

```python
nums=[1, 2, 3, 2, 3, 1, 3]
print(one_odd_occuring(nums=nums)) # time complexity : O(n)
```

```python
# mehtod 2
def one_odd_occuring(nums:list):
    odd_val=0
    for i in nums:
        odd_val=odd_val^i
    return odd_val
```

```python
nums = [1, 2, 3, 2, 3, 1, 3]
print(one_odd_occuring(nums=nums))  # time complexity : O(n)
```
