```python
def immediateSmaller(nums:list, x:int):
    ans=-1
    diff=10000000
    for i in nums:
        if(i<x and x-i<diff):
            ans=i
            diff=x-i
    return ans
```

```python
nums = [63, 6, 60, 38, 3, 94, 93, 83, 65]
x = 9
print(immediateSmaller(nums=nums, x=x)) # time complexity: o(n)
```
