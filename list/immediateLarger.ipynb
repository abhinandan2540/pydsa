```python
def immediateLarger(nums:list, x:int):
    ans=-1
    diff=10000000
    for i in nums:
        if(i>x and i-x<diff):
            ans=i
            diff=i-x
    return ans
```

```python
nums = [63, 6, 60, 38, 3, 94, 93, 83, 65]
x=9
print(immediateLarger(nums=nums,x=x)) # time complexity: O(n)
```
