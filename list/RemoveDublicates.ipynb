```python
# method 1
def remove_dublicates(nums:list):
    new_nums=[]
    for i in nums:
        if i not in new_nums:
            new_nums.append(i)
    return new_nums

```

```python
nums = [1,2,3,3,4]
print(remove_dublicates(nums=nums)) # time complexity: O(n)
```

```python
# method 2
def remove_dublicates(nums: list):
    for i in nums:
        if nums.count(i)>1:
            nums.remove(i)
    return nums
```

```python
nums = [1, 2, 3, 3, 4]
print(remove_dublicates(nums=nums))  # time complexity: O(n)
```
