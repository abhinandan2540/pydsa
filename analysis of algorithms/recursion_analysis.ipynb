```python
def recursive_1(n: int):
    if (n == 1):
        print('dsa last')
        return
    else:
        print('dsa')
    return recursive_1(n-1)
```

```python
recursive_1(5) # time complexity: O(n)
```

```python
def recursion_2(n: int):
    if (n == 1):
        print('dsa end')
        return
    for _ in range(n):
        print('dsa')
    recursion_2(n-1)
    recursion_2(n-1)
    
```

```python
recursion_2(5) # time complexity: O(n^2)
```
