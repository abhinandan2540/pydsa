```python
def multiple_while_loops(n:int):
    i=0
    while(i<n): # O(n)
        print(i)
        i+=2

    j=2
    while(j<n): # o(logn)
        print(j)
        j*=2
    
    k=1
    while(k<n): # O(n)
        print(k)
        k+=1
    
```

```python
multiple_while_loops(10)
# time complexity: O(n)
```
