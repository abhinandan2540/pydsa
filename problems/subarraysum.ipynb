```python
# method 1
# for subarray sum, applying bruteforce approach
def subarray_sum(arr:list, target:int):
    for i in range(len(arr)):
            arr_sum=0
            for j in range(i,len(arr)):
                arr_sum+=arr[j]
                if(arr_sum==target):
                    return list((i+1,j+1))
    return -1
```

```python
arr=[1,2,3,7,5]
target=12
print(subarray_sum(arr=arr, target=target))
# time complexity: O(n^2)
```

```python
# method 2
# sliding window approach
def subarray_sum(arr:list, target:int):
    curr_sum=0
    j=0

    for i in range(len(arr)):
        curr_sum+=arr[i]

        while curr_sum>target and j<=i:
            curr_sum-=arr[j]
            j+=1

        if (curr_sum==target):
            return [j+1, i+1]
    return -1
```

```python
arr=[1,2,3,7,5]
target=12
print(subarray_sum(arr=arr, target=target))
# time complexity: O(n)
```
